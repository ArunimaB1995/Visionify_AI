```python
import datetime
import calendar
import re
# get current time
now = datetime.datetime.now()
try:
    with open('log1.log', 'r') as f:
        # go through each line
        for ln in f:
            # ex: 'Sep 8 13:25:02 etcetcetc'
            stime = ln.split(' ')
            completeTime = stime[3]+ ' ' + stime[4]
            datefstr = datetime.datetime.strptime(completeTime, '[%d/%b/%Y:%H:%M:%S %z]')
    #         mic_sec = dat_time.microsecond
    #         timcon = calendar.timegm(dat_time.timetuple())*1000000 + mic_sec
            if (now.hour - datefstr.hour) < 1 and (now.minute - datefstr.minute) < 60:
                print(datefstr)
            else: 
                continue 
except:
    found = re.match('2015', completeTime)
    if not found:
        print("time data %r does not match format %r" %(datefstr, now))
```

```python
import datetime
import calendar
import re
from collections import Counter

# get current time
count = 0
with open('log1.log', 'r') as f:
    # go through each line
    for ln in f:
        ip = ln.split(' ')
        ip_add = ip[0]+ ' '
        if re.match('', ip_add) == None:
            print("None")
        else:
            count+=1
            print(ip_add)
```

```python
from pprint import pprint
import os
requestCount = {}

data = open('log1.log', 'r').readlines()

for line in data:
    lineParts = line.split(" ")
    ip = lineParts[0] + ' '
    method = lineParts[2]

    if ip not in requestCount:
        requestCount[ip] = {}

    if method not in requestCount[ip]:
        requestCount[ip][method] = 0
    else:    
        requestCount[ip][method] += 1

pprint(requestCount)
```
